```python
import sqlite3
con = sqlite3.connect("lahman_1871-2022.sqlite")
```

```python
cursor = con.cursor()
get_schema='''
        select *
        from sqlite_schema
        where type="table";
        '''
cursor.execute(get_schema)
result=cursor.fetchall()
cursor.close()
```

```python
import pandas as pd
lahman_df=pd.read_sql(get_schema,con)
lahman_df
```

```python
pd.read_sql("select * from Pitching",con)
```

```python
pd.read_sql("select * from Batting",con)
```

```python
pd.read_sql("select * from HallOfFame where inducted='Y'",con)
```

```python
inner_1="""
    select distinct P.playerID,sum(P.GS),sum(P.G),sum(P.W),sum(P.L),sum(IPOuts),sum(CG),sum(SHO),sum(SV) from Pitching as P
    inner join HallOfFame as H on H.playerID=P.playerID
    where H.inducted='Y'
    group by P.playerID
    """
HOF_Pitching_Stats=pd.read_sql(inner_1,con)
HOF_Pitching_Stats
```

```python
inner_2="""
    select distinct B.playerID,sum(AB),sum(R),sum(H),sum(HR),sum(RBI),sum(BB),sum(SO) from Batting as B
    inner join HallOfFame as H on H.playerID=B.playerID
    where H.inducted='Y'
    group by B.playerID
    """
HOF_Batting_Stats=pd.read_sql(inner_2,con)
HOF_Batting_Stats
```
